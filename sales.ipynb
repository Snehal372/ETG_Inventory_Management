# Purchase

```python
import time
import uuid
print("                LEGAL RECEIPT              \n")
print("****************************************************")
print("         BHALE'S Grocery Store                ")
print("****************************************************")

m=0
print(time.ctime())
name=str(input("Name :"))
print("Transaction ID:",end=" ")
UI = uuid.uuid1()
print(UI)
print("====================================================")
n =int(input("\nEnter how many products you want to order:"))
for i in range (n):
            user_prod=str(input("Enter the product ID: "))
            user_quant=int(input("Enter the quantity of product: "))
            print("Product:", records[user_prod]['name'])
            print("price:",records[user_prod]['cost'])    
            print("====================================================")
            print("Amount :",(records[user_prod]['cost']*(user_quant)))
            m= m+((records[user_prod]['cost'])*(user_quant))
            print("====================================================\n")
       

print("----------------------------------------------------")
print("Total Amount :",m)
print("----------------------------------------------------")

if(m>0):
    if m<=5000:
        disc = m*0.05
    elif m<=15000:
        disc=m*0.12
    elif m<=25000:
        disc=0.2 * m
    else:
        disc=0.3 * m
    
    print("----------------------------------------------------")

    print("YOUR DISCOUNT IS : ",disc)
    print("YOU HAVE TO PAY  : ",m-disc)
    print("----------------------------------------------------")
else:
    print("Invalid Amount")

print("          Thank you for Shopping :)           ")


records[user_prod]['Quantity'] = records[user_prod]['Quantity'] - (int(user_quant))
```

# Update Inventory after purchase

```python
js = json.dumps(records)

fd = open("records.json",'w')
fd.write(js)
fd.close()
```
