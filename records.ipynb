# Read Data From Json

```python
import json
```

```python
fd = open("records.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)
```

```python
js = json.dumps(records)
```

```python
fd = open("records.json",'w')

fd.write(js)

fd.close()
```

# Add New Item into Inventory

```python
n =int(input("\nEnter how many products you want to add into the inventory:"))
for i in range (n):  
    prod_id = str(input("\nEnter product ID:"))
    name = str(input("Enter name:"))
    sr_no = int(input("Enter serial number :"))
    cost = int(input("Enter price:"))
    Type = str(input("Enter the Type of product :"))
    Quantity = str(input("Enter quantity:"))
    DOM =str(input("Date of Manufacturing :"))
    Exp_Date =str(input("Exp Date:"))

    records[prod_id] = {"Type":Type,'Sr':sr_no, 'name': name, 'cost': cost, 'Quantity':Quantity,
                    'Date':{ 'DOM': DOM,'Exp_Date': Exp_Date}}

js = json.dumps(records)

fd = open("records.json",'w')
fd.write(js)
fd.close()
```

# Updating quantity of products

```python
c =int(input("how many product's quantity you wanna add?"))
for i in range(c):
    prod_id = str(input("\nEnter product ID:"))
    qt = int(input("Enter the quantity of products you want to add:"))
    records[prod_id]['Quantity']= qt+int(records[prod_id]['Quantity'])
    print("Your quantity has been updated!!")
    print("The Quantity after addition is:" + str(records[prod_id]["Quantity"]) )
```
